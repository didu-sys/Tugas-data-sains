```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Load data
df = pd.read_csv('emails.csv')

# 2. Tentukan fitur (x) dan target (y)
# Kita gunakan kolom 'the' sebagai contoh fitur (x) 
# Dan kolom 'Prediction' sebagai target (y)
x = df[['the']] 
y = df['Prediction']

# 3. Latih model
model = LinearRegression()
model.fit(x, y)

# 4. Visualisasi
plt.scatter(x, y, color='blue', label='Data Asli')
plt.plot(x, model.predict(x), color='red', linewidth=2, label='Garis Regresi')
plt.title('Frekuensi Kata "the" vs Prediksi Spam')
plt.xlabel('Jumlah Kata "the"')
plt.ylabel('Spam (1) atau Bukan (0)')
plt.legend()
plt.show()

# 5. Prediksi data baru
# Contoh: Jika kata "the" muncul 5 kali
new_data = pd.DataFrame({'the': [5]})
prediksi = model.predict(new_data)[0]
print(f"Hasil prediksi nilai spam untuk 5 kata 'the' adalah: {prediksi}")
```
